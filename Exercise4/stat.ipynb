```python
from math import factorial, exp
import matplotlib.pyplot as plt
import seaborn as sns
def P(x):
    return (((pow(10,x))/(factorial(x)))*exp(-10))

x=[i for i in range(1,11)]
y=[P(j) for j in range(1,11)]
sns.set_style("whitegrid")
r = sns.color_palette("RdPu", 1)
sns.set_palette(r)
sns.scatterplot(x=x,y=y,s=100)
plt.xlabel("x")
plt.ylabel("p(x)")
plt.show()

print("P(X = 10) = "+ str(P(10))+ "\n")
print("P(X <= 10) = "+ str(sum(y))+ "\n")
```
